Bayes' Theorem is expressed as :----:  
P(A|B) = [P(B|A) * P(A)] / P(B)

**Type**

1. Gussians naive bayes

    Used when features are continuous and follow a normal (Gaussian) distribution.

    Example: predicting height, weight, or exam scores.

2. Multinomial Naïve Bayes

    Used for discrete count features (like word frequency in text).

    Example: text classification, spam detection, document categorization.

3. Bernoulli Naïve Bayes

    Used for binary/boolean features (0 or 1).

    Example: whether a word appears in an email (yes/no)


```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv('/content/spam.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Category'].value_counts()
```

```python
df['Category']=df['Category'].map({'spam':1,'ham':0})
```

```python
df.head()
```

```python
x=df['Message']
y=df['Category']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Vectoeizer

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([
    ('vector',CountVectorizer()),
    ('model',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
pred=clf.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred))
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
```

```python
clf.predict(emails)
```
