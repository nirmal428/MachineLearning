```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.boxplot(data=df,ax=ax)
```

```python
x=df.drop("species",axis=1)
y=df["species"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model , filled=True)
```

post prunning

```python
# post prunning eith max_depth =2
model=DecisionTreeClassifier(max_depth=2)
model.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model , filled=True)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Decision Tree with Pre-Pruning

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
parameter={
    'criterion':['gini','entropy'],
    'max_depth':[4,5,6,7],
    'splitter':['best','random'],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier=DecisionTreeClassifier()
cv=GridSearchCV(treeclassifier,param_grid=parameter,cv=5,scoring="accuracy")
cv.fit(x_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred=cv.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Decision Tree Regressor

```python
from sklearn.datasets import fetch_california_housing
california_df=fetch_california_housing()
```

```python
california_df
```

```python
x=pd.DataFrame(california_df.data,columns=california_df.feature_names)
y=california_df.target
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```

```python
score
```

```python
parameter={
    'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
    'splitter':['best','random'],
    'max_depth':[4,5,6,7],
    'max_features':['auto','sqrt','log2']
}
```

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import GridSearchCV
cv=GridSearchCV(model,param_grid=parameter,cv=5,scoring="neg_mean_squared_error")
cv.fit(x_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred=cv.predict(x_test)
y_pred
```

```python
score=r2_score(y_test,y_pred)
score
```
