**EDA (Exploratory Data Analysis) ->** EDA is the process of analyzing datasets to understand their structure, detect patterns, identify anomalies, and gain insights before modeling. It involves:

**Data Cleaning ->** Handling missing or inconsistent data.

**Descriptive Statistics ->** Mean, median, standard deviation, etc.

**Univariate & Multivariate Analysis ->** Studying variables individually and together.

**Visualization ->**Using histograms, box plots, scatter plots, and heatmaps.
__________

It helps in better decision-making for data preprocessing and model selection.
_____________

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

DataLinke = https://www.kaggle.com/datasets/nikhilmittal/flight-fare-prediction-mh?select=Data_Train.xlsx

```python
df=pd.read_excel("/content/Data_Train.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(2)
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_hours']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df['Arrival_hours']=df['Arrival_hours'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```
