# Support Vector Machines Classifier  

**Support Vector Machine (SVM) Classifier** is a supervised learning algorithm that classifies data by finding the optimal hyperplane that maximizes the margin between classes. It uses **support vectors** (closest data points) and can handle linear or non-linear data using **kernels**.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,
                        n_redundant=0,n_classes=2,n_clusters_per_class=2)
print(x.shape,y.shape)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear',random_state=0)
```

```python
svc.fit(x_train,y_train)
```

```python
svc.coef_
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

**#  Hyperparametere Tuning With SVC**

```python
from sklearn.model_selection import GridSearchCV
```

```python
para_grid={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid=GridSearchCV(SVC(),para_grid,refit=True,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
