# Naive Bayes

Naive Bayes is a simple and fast classification algorithm based on Bayes’ Theorem. It assumes that features are independent of each other (naive assumption).

Key Points:

Probabilistic model – predicts the class with the highest posterior probability.

Works well with text classification, spam filtering, sentiment analysis.

Types: Gaussian, Multinomial, Bernoulli Naive Bayes.

Advantages: Fast, works with small datasets, handles high-dimensional data well.

Limitation: Assumes feature independence, which may not always hold true.

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
x,y=load_iris(return_X_y=True)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
print(print(classification_report(y_test,y_pred)))
print(confusion_matrix(y_test,y_pred))
```
